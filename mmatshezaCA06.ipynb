# Perform K means clustering

## Task 2

Prepare the data for clustering: a. Perform any necessary feature scaling 
(StandardScaler or MinMaxScaler) b. Choose the appropriate features for 
clustering (you may start with 'Annual Income' and 'Spending Score') c. Create a 
new DataFrame with only the selected features

```python
# Task 2

# a. Perform any necessary feature scaling (StandardScaler or MinMaxScaler)
from sklearn.preprocessing import StandardScaler


```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])

# b. Choose the appropriate features for clustering
# (you may start with 'Annual Income' and 'Spending Score')
selected_features = ['Annual Income (k$)', 'Spending Score (1-100)']

# c. Create a new DataFrame with only the selected features
scaled_data_df = pd.DataFrame(scaled_data, columns=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])
cluster_data = scaled_data_df[selected_features]
```

```python
scaled_data_df.head()
```

```python
cluster_data.head()
```

## Task 3

Implement k-means clustering: a. Import the KMeans class from the sklearn.cluster module b. Use the Silhouette Method to determine the optimal number of clusters c. Train the KMeans model with the optimal number of clusters d. Obtain the cluster assignments for each data point.

```python
# Task 3

# a. Import the KMeans class from the sklearn.cluster module
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

```

```python
# b. Use the Silhouette Method to determine the optimal number of clusters
n_clusters_range = range(2, 11)
silhouette_scores = []

for n_clusters in n_clusters_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(cluster_data)
    silhouette_avg = silhouette_score(cluster_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

optimal_n_clusters = n_clusters_range[silhouette_scores.index(max(silhouette_scores))]
print(f"Optimal number of clusters: {optimal_n_clusters}")

# c. Train the KMeans model with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=42)
kmeans.fit(cluster_data)

# d. Obtain the cluster assignments for each data point
cluster_assignments = kmeans.predict(cluster_data)
```

## Task 4

Visualize and analyze the clusters: a. Create a scatter plot of the selected features, 
colored by cluster assignment b. Interpret the clusters and provide a brief 
description of each cluster c. (Optional) Perform the same analysis with different 
sets of features and compare the result

```python
# Task 4

# a. Create a scatter plot of the selected features, colored by cluster assignment
import matplotlib.pyplot as plt

plt.scatter(cluster_data.iloc[:, 0], cluster_data.iloc[:, 1], c=cluster_assignments, cmap='viridis', marker='o', s=50)
plt.xlabel(cluster_data.columns[0])
plt.ylabel(cluster_data.columns[1])
plt.title(f"KMeans Clustering (k={optimal_n_clusters})")
plt.colorbar().set_label("Cluster")
plt.show()
```

### Process repeated with features 'Age and Spending Score.'

```python
# Create a new DataFrame with different set of features
new_cluster_data = data[['Age', 'Spending Score (1-100)']]

# Normalize the new data
new_cluster_data = scaler.fit_transform(new_cluster_data)

# Find the optimal number of clusters for the new data
new_silhouette_scores = []

for n_clusters in n_clusters_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(new_cluster_data)
    silhouette_avg = silhouette_score(new_cluster_data, cluster_labels)
    new_silhouette_scores.append(silhouette_avg)

new_optimal_n_clusters = n_clusters_range[new_silhouette_scores.index(max(new_silhouette_scores))]
print(f"Optimal number of clusters for new features: {new_optimal_n_clusters}")

# Train the KMeans model with the optimal number of clusters for the new data
kmeans = KMeans(n_clusters=new_optimal_n_clusters, random_state=42)
kmeans.fit(new_cluster_data)

# Obtain the cluster assignments for each data point in the new data
new_cluster_assignments = kmeans.predict(new_cluster_data)

# Plot the new clusters
plt.scatter(new_cluster_data[:, 0], new_cluster_data[:, 1], c=new_cluster_assignments, cmap='viridis', marker='o', s=50)
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title(f"KMeans Clustering for new features (k={new_optimal_n_clusters})")
plt.colorbar().set_label("Cluster")
plt.show()
```
